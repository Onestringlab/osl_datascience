#Soal
Sesorang memiliki sekantong permen. Terdapat lima jenis permen, masing-masing diberi
nama di bawah ini. Buat diagram yang menunjukkan persentase peluang bahwa akan
mengeluarkan permen Snickers dari kantong jika dilakukan pengambilan acak.
Sebutkan peluang memilih permen Snickers.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as npt 
```

```python
candy_names = ['Kit Kat', 'Snickers', 'Milky Way', 'Toblerone', 'Twix']
candy_counts = [52, 39, 90, 13, 78]
colors = ('#8B4513', '#FFF8DC', '#93C572', '#E67F0D', '#D53032')
explode = (0, 0, 0.1, 0, 0)
```

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.pie(candy_counts, labels = candy_names)
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.pie(candy_counts, labels = candy_names, autopct='%1.0f%%')
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.pie(candy_counts, labels = candy_names, autopct='%1.0f%%')
ax.set_title('Diagram Permen Favorit')
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.pie(candy_counts, labels = candy_names, autopct='%1.0f%%', colors=colors)
ax.set_title('Diagram Permen Favorit')
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.pie(candy_counts, labels = candy_names, autopct='%1.0f%%', colors=colors
       ,explode=explode)
ax.set_title('Diagram Permen Favorit')
```
